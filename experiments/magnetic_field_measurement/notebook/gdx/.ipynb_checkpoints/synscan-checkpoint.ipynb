```python
import serial
import serial.tools.list_ports
import time
```

```python
com_ports=serial.tools.list_ports.comports()
for p in com_ports:
    print(p.device)
```

```python
skywatcher = serial.Serial('/dev/ttyUSB1', baudrate=115200, timeout=0.02)
```

```python
msg=bytes(f':{'j'}{2}\r','utf-8')
```

```python
skywatcher.write(msg)
```

```python
while True:
    skywatcher.write(msg)
    time.sleep(0.1)
    response= skywatcher.read_until()
    print(response.decode())
```
